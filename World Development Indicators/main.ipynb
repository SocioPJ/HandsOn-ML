```python
i=0
nome_colunas = []
for colunas in dataset.columns:
  nome_colunas.append(colunas.replace(' ', ''))
  dataset = dataset.rename(columns={colunas: colunas.replace(' ', '')})
  i += 1
  if i == len(dataset.columns) - 1:
    print(nome_colunas)
```
